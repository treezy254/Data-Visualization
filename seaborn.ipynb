## Line Charts

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
spotify_data = pd.read_csv('spotify.csv', index_col='Date', parse_dates=True)
spotify_data.head()
```

```python
plt.figure(figsize=(14,6))
plt.title('Daily Global Stream of Popular Song from 2017 - 2018')
sns.lineplot(data=spotify_data)
```

```python
list(spotify_data.columns)
```

```python
plt.figure(figsize=(14,6))
plt.title('Daily Global Stream of Popular Songs in 2017 - 2018')
sns.lineplot(data=spotify_data['Shape of You'], label='Shape of You')
sns.lineplot(data=spotify_data['Despacito'], label='Despacito')
plt.xlabel('Date')
```

## Bar Charts and Heatmaps

```python
flight_data = pd.read_csv('flight_delays.csv', index_col='Month')
flight_data.head()
```

```python
plt.figure(figsize=(10,6))
plt.title('Average Arrival Delay for Spirit Airlines Flights, by Month')

sns.barplot(x=flight_data.index, y=flight_data['NK'])
plt.ylabel('Arrival delay (in minutes')
```

### Heatmap

```python
plt.figure(figsize=(14,6))
plt.title('Average Arrival Delay for Each Airline (by month')
sns.heatmap(data=flight_data, annot=True)
plt.xlabel('Airline')
```

## Scatter Plots

```python
insurance_data = pd.read_csv('insurance.csv', index_col=0)
insurance_data.head()
```

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

```python
sns.lmplot(x='bmi', y='charges', hue='smoker', data=insurance_data)
```

```python
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges'])
```

## Distributions

```python
iris_data = pd.read_csv('iris.csv', index_col=0)
iris_data.head()
```

### Histograms

```python
sns.histplot(iris_data['Petal Length (cm)'])
```

### Density plots

```python
# kernel density plot
sns.kdeplot(data=iris_data['Petal Length (cm)'], shade=True)
```

### 2D KDE Plots

```python
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind='kde')
```

## Color coded plots

```python
sns.histplot(data=iris_data, x='Petal Length (cm)', hue='Species')
plt.title('Histogram of Petal Lengths, by Species')
```

```python
sns.set_style('dark')
sns.kdeplot(data=iris_data, x='Petal Length (cm)', hue='Species', shade=True)
plt.title('Histogram of Petal Lengths, by Species')
```
